```python
import pandas as pd
import numpy as np

# ========= 1. Загрузка =========
file_path = "grimms_fairytales.csv"   # <-- поменяй на свой файл

if file_path.endswith(".csv"):
    df = pd.read_csv(file_path)
elif file_path.endswith(".xlsx") or file_path.endswith(".xls"):
    df = pd.read_excel(file_path)
else:
    raise ValueError("Неподдерживаемый формат файла")

print("\n===== SHAPE =====")
print(df.shape)

# ========= 2. Первичный просмотр =========
print("\n===== HEAD =====")
print(df.head())

print("\n===== SAMPLE =====")
print(df.sample(5))

print("\n===== INFO =====")
print(df.info())

# ========= 3. Типы данных =========
print("\n===== DTYPES =====")
print(df.dtypes)

# ========= 4. Пропуски =========
print("\n===== MISSING VALUES =====")
missing = df.isnull().sum()
print(missing[missing > 0].sort_values(ascending=False))

print("\nПроцент пропусков:")
print((df.isnull().mean() * 100).sort_values(ascending=False))

# ========= 5. Дубликаты =========
print("\n===== DUPLICATES =====")
print("Количество дубликатов:", df.duplicated().sum())

# ========= 6. Описательная статистика =========
print("\n===== NUMERIC DESCRIPTION =====")
print(df.describe())

print("\n===== ALL COLUMNS DESCRIPTION =====")
print(df.describe(include='all'))

# ========= 7. Уникальные значения (для категорий) =========
print("\n===== UNIQUE VALUES (categorical) =====")
cat_cols = df.select_dtypes(include=['object', 'category']).columns

for col in cat_cols:
    print(f"\n--- {col} ---")
    print("Unique count:", df[col].nunique())
    print(df[col].value_counts().head(10))

# ========= 8. Корреляция =========
print("\n===== CORRELATION MATRIX =====")
numeric_df = df.select_dtypes(include=np.number)
if not numeric_df.empty:
    print(numeric_df.corr())
else:
    print("Нет числовых колонок")

# ========= 9. Быстрая проверка распределений =========
print("\n===== BASIC DISTRIBUTION CHECK =====")
for col in numeric_df.columns:
    print(f"\n--- {col} ---")
    print("Min:", df)
```
